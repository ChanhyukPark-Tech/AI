```python
# Use library tensorflow and pandas
import tensorflow as tf
import pandas as pd
```

```python
# prepare previous data for our purpose
location = 'C:/Users/woner/Desktop/coding/machine_learning/(me)life_coding_Tensorflow101/csv/iris.csv'
iris = pd.read_csv(location)
iris.head()
```

```python
# we have to do one hot encoding because not regression but classification
encoding = pd.get_dummies(iris)
encoding.head()
print(encoding.columns)
```

```python
# independent variable , dependent variable 
independent = encoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
dependent = encoding[['품종_setosa', '품종_versicolor',
       '품종_virginica']]
print(independent.shape , dependent.shape)
```

```python
# model structure
X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation='softmax')(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy',
             metrics = 'accuracy')
```

```python
# Fit  & learn  !
model.fit(independent,dependent,epochs = 500 )
```

```python
# use our model
model.predict(independent[-5:])
```
